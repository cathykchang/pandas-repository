```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

```python
# list out columns in df
list(school_data_complete)
```

```python
# find total unique school names
number_of_schools = school_data_complete['school_name'].nunique()
number_of_schools                   
```

```python
# find total unique student names
total_students = len(school_data_complete['student_name'])
total_students
```

```python
# sum total budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
# find average math score using mean fx
ave_math_score = school_data_complete['math_score'].mean()
ave_math_score
```

```python
# find average reading score using mean fx
ave_reading_score = school_data_complete['reading_score'].mean()
ave_reading_score
```

```python
# use conditional to locate math scores >=70 and count the passing math students
students_passing_math = school_data_complete.loc[school_data_complete['math_score']>=70]['math_score'].count()
students_passing_math
```

```python
# calculate % of students passing math
p_passing_math = students_passing_math/total_students
p_passing_math
```

```python
# use conditional to locate reading scores >=70 and count the passing reading students
students_passing_reading = school_data_complete.loc[school_data_complete['reading_score']>=70]['reading_score'].count()
students_passing_reading
```

```python
# calculate % of students passing reading
p_passing_reading = students_passing_reading/total_students
p_passing_reading
```

```python
# use conditional to locate reading and math scores >=70 and count the passing overall students
students_passing_overall = school_data_complete.loc[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)]['student_name'].count()
students_passing_overall               
```

```python
# calculate % of students passing overall
p_passing_overall = students_passing_overall / total_students
p_passing_overall
```

```python
# print dataframe with results
district_summary_df = pd.DataFrame({
    "Total Schools": [number_of_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [ave_math_score],
    "Average Reading Score": [ave_reading_score],
    "% Passing Math": p_passing_math*100,
    "% Passing Reading": p_passing_reading*100,
    "% Overall Passing": p_passing_overall*100
})
district_summary_df
```

```python
# format budget to $
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# populate ranges for bins and names for labels  
spending_bins = [0, 585, 630, 645, 680]
ranges = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# calculate per student spend
per_student_spend = total_budget_per_school / groupednewsize
school_summary_df['Per Student Spend'] = per_student_spend

school_summary_df
```

```python
# apply cut function to sort per student spend into bins
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Spend"], bins=spending_bins, labels=["<$585", "$585-630", "$630-645", "$645-680"], include_lowest=True)
school_summary_df
```

```python
# calculate per spending range bin averages using groupby and mean fx
spending_math_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
# set spending ranges (per student) column as index
spend_index = school_summary_df.set_index("Spending Ranges (Per Student)")
spend_index
```

```python
# print results in dataframe
spending_summary_df = pd.DataFrame({    
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})
spending_summary_df
```

```python
# format results with 2 decimal places
spending_summary_df["Average Math Score"] = spending_summary_df["Average Math Score"].astype(float).map("{:.2f}".format)
spending_summary_df["Average Reading Score"] = spending_summary_df["Average Reading Score"].astype(float).map("{:.2f}".format)
spending_summary_df["% Passing Math"] = spending_summary_df["% Passing Math"].astype(float).map("{:.2f}".format)
spending_summary_df["% Passing Reading"] = spending_summary_df["% Passing Reading"].astype(float).map("{:.2f}".format)
spending_summary_df["% Overall Passing"] = spending_summary_df["% Overall Passing"].astype(float).map("{:.2f}".format)
spending_summary_df
```
